```python
df = pd.read_csv("/content/sample_data/indorelawan.csv", sep=",", encoding='utf8', engine="python")

df.dropna(inplace=True)
```

```python
from sklearn import preprocessing

df.rename(columns = {'keahlian 1':'keahlian_1', 'keahlian 2':'keahlian_2', 'nomor': 'foundation_id'}, inplace = True)
le = preprocessing.LabelEncoder()
le.fit(df.keahlian_1)
df['keahlian_1'] = le.transform(df.keahlian_1)
df['keahlian_2'] = le.transform(df.keahlian_2)

df
```

```python
#creating dummy dataset for training

def create_dataframe(volunteerid, le):
  import random

  volunteer_id = []
  count = []

  for i in range(19):
    volunteer_id.append(volunteerid)

  for i in range(19):
    count.append(random.randint(0, 10))

  list_features = pd.DataFrame(volunteer_id)
  list_features.columns = ["volunteer_id"]
  list_features["keahlian"] = list(le.classes_)
  list_features["pengalaman"] = count

  return list_features

def add_dataframe(dataframe1, dataframe2):
  dataframe = pd.concat([dataframe1, dataframe2], axis=0, ignore_index=True)
  return dataframe

for i in range(21):
  if i % 2 == 0:
    continue
  else:
    one = create_dataframe(i, le)
    two = create_dataframe(i+1, le)
    three = add_dataframe(one, two)
    if i == 1:
      dummy = three
    else:
      dummy = add_dataframe(dummy, three)

dummy
```

```python
#removing random rows to make the data more "real"
dummy = dummy[dummy['pengalaman'] > 0]

remove_n = 100
drop_indices = np.random.choice(dummy.index, remove_n, replace=False)
dummy_subset = dummy.drop(drop_indices)

dummy_subset.reset_index(inplace=True)

dummy_subset.drop(columns=["index"], inplace=True)

dummy_subset
```

```python
#building the model

import tensorflow as tf
import tensorflow_recommenders as tfrs
import keras 
from datetime import datetime

dummy_subset['volunteer_id'] = dummy_subset.volunteer_id.astype(np.str)
dummy_subset['keahlian'] = dummy_subset.keahlian.astype(np.str)
dummy_subset['pengalaman'] = dummy_subset.pengalaman.astype(np.float32)

dataset = tf.data.Dataset.from_tensor_slices((tf.cast(dummy_subset['volunteer_id'].values.reshape(-1,1), tf.string), tf.cast(dummy_subset['keahlian'].values.reshape(-1,1), tf.string),
tf.cast(dummy_subset['pengalaman'].values.reshape(-1,1),tf.float32)))

@tf.function
def rename(x0,x1,x2):
    y = {}
    y["volunteer_id"] = x0
    y['keahlian'] = x1
    y['pengalaman'] = x2
    return y

dataset = dataset.map(rename)

keahlian = dummy_subset.keahlian.values
volunteer = dummy_subset.volunteer_id.values

unique_keahlian = np.unique(list(keahlian))
unique_volunteer_ids = np.unique(list(volunteer))
```

```python
class RankingModel(tf.keras.Model):

  def __init__(self):
    super().__init__()
    embedding_dimension = 5

    # Compute embeddings for users.
    self.volunteer_embeddings = tf.keras.Sequential([
      tf.keras.layers.experimental.preprocessing.StringLookup(
        vocabulary=unique_volunteer_ids, mask_token=None),
      tf.keras.layers.Embedding(len(unique_volunteer_ids) + 1, embedding_dimension)
    ])

    # Compute embeddings for books.
    self.keahlian_embeddings = tf.keras.Sequential([
      tf.keras.layers.experimental.preprocessing.StringLookup(
        vocabulary=unique_keahlian, mask_token=None),
      tf.keras.layers.Embedding(len(unique_keahlian) + 1, embedding_dimension)
    ])

    # Compute predictions.
    self.ratings = tf.keras.Sequential([
      # Learn multiple dense layers.
      tf.keras.layers.Dense(16, activation="relu"),
      tf.keras.layers.Dense(32, activation="relu"),
      # Make rating predictions in the final layer.
      tf.keras.layers.Dense(1)
  ])

  def __call__(self, x):
    
    volunteer_id, keahlian = x
    volunteer_embedding = self.volunteer_embeddings(volunteer_id)
    keahlian_embedding = self.keahlian_embeddings(keahlian)

    return self.ratings(tf.concat([volunteer_embedding, keahlian_embedding], axis=1))
```

```python
class volunteersModel(tfrs.models.Model):

  def __init__(self):
    super().__init__()
    self.ranking_model: tf.keras.Model = RankingModel()
    self.task: tf.keras.layers.Layer = tfrs.tasks.Ranking(
      loss = tf.keras.losses.MeanSquaredError(),
      metrics=[tf.keras.metrics.RootMeanSquaredError()]
    )

  def compute_loss(self, features, training=False) -> tf.Tensor:
    print(features)
    mission_predictions = self.ranking_model((features['volunteer_id'], features["keahlian"]))

    # The task computes the loss and the metrics.
    return self.task(labels=features["pengalaman"], predictions=mission_predictions)
```

```python
model = volunteersModel()
model.compile(optimizer=tf.keras.optimizers.Adagrad(learning_rate=0.1))
# Cache the dataset 
cache_dataset = dataset.cache()
# Tensorboard 
logdir = "logs/scalars/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=logdir)
# Training 
model.fit(cache_dataset, epochs=25,
                    verbose=1, callbacks=[tensorboard_callback])
```

```python
# Create array with users id in every place on the lenght of the unique books number
user8 = np.array(["8" for i in range(len(unique_keahlian))])

# Convert it to tf.data.Dataset 
test_data = tf.data.Dataset.from_tensor_slices((tf.cast(user8.reshape(-1,1), tf.string), tf.cast(unique_keahlian.reshape(-1,1), tf.string)))
# Name the columns 
@tf.function
def rename_test(x0,x1):
    y = {}
    y["volunteer_id"] = x0
    y['keahlian'] = x1
    return y
test_data = test_data.map(rename_test)
# Now lets make predictions and store them in to dictionary
test_mission = {}
for b in test_data:
    test_mission[b['keahlian'].numpy()[0]] = model.ranking_model((b['volunteer_id'],b['keahlian']))
# sort them by score and print the titles 
for b in sorted(test_mission, key=test_mission.get, reverse=True):
    print(b)
```
